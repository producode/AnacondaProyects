```python
def sectorToInt(sector):
    if sector == "Industrials":
        return 1
    if sector == "Health Care":
        return 2
    if sector == "Information Technology":
        return 3
    if sector == "Communication Services":
        return 4
    if sector == "Consumer Staples":
        return 5
    if sector == "Consumer Discretionary":
        return 6
    if sector == "Utilities":
        return 7
    if sector == "Financials":
        return 8
    if sector == "Materials":
        return 9
    if sector == "Real Estate":
        return 10
    if sector == "Energy":
        return 11
    return 0
print(sectorToInt("Communication Services"))
```
